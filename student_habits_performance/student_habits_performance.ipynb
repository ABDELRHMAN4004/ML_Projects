```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False)
encoded_gender=encoder.fit_transform(file[['gender']])
encoded_gender=pd.DataFrame(encoded_gender,columns=encoder.get_feature_names_out(['gender']))
file=pd.concat([file,encoded_gender],axis=1)
file.drop(columns=['gender'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False)
encoded_dite=encoder.fit_transform(file[['diet_quality']])
encoded_dite=pd.DataFrame(encoded_dite,columns=encoder.get_feature_names_out(['diet_quality']))
file=pd.concat([file,encoded_dite],axis=1)
file.drop(columns=['diet_quality'],axis=1,inplace=True)
```

# # Show relashonship between features

```python
plt.figure(1,(17,8))
sns.heatmap(file.corr(),linewidths=.3,annot=True,fmt='.1g')
plt.show()
```

# make scaler

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_scale = ['age', 'mental_health_rating', 'exercise_frequency', 'sleep_hours','attendance_percentage', 'netflix_hours', 'social_media_hours', 'study_hours_per_day']
file[columns_to_scale] = scaler.fit_transform(file[columns_to_scale])

```

```python
file
```

# Define features and labels

```python
labels=file['exam_score']
features=file.drop(columns=['exam_score'],axis=1)
```

# Split data to train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.3,random_state=100)
```

# Apply XGBoost

```python
import xgboost as xgb

model_1=xgb.XGBRegressor( n_estimators=200,learning_rate=0.05,max_depth=5,subsample=0.8,colsample_bytree=0.8,random_state=42)

model_1.fit(x_train,y_train)
```

```python
model_1_predictions=model_1.predict(x_test)
```

# Show error and accuracy of model

```python
from sklearn.metrics import r2_score,root_mean_squared_error
acc = r2_score(y_test, model_1_predictions)
rmse=root_mean_squared_error(y_test, model_1_predictions)
print(f"rmse: {rmse}")
print(f"Accuracy: {acc }")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8,6))
plt.scatter(y_test, model_1_predictions, color='blue', label='Predicted values Vs True values')
plt.plot(y_test, y_test, color='red', linestyle='--', label='best relashonship(y = x)')
plt.xlabel('True values')
plt.ylabel('Predicted values')
plt.title('Predicted values Vs True values')
plt.legend()
plt.grid(True)
plt.show()
```
