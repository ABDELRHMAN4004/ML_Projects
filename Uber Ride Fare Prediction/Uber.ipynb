```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file.dropna(inplace=True)
```

```python
file.drop(columns='Unnamed: 0',inplace=True)
```

```python
file['pickup_datetime']=pd.to_datetime(file['pickup_datetime'])
```

```python
file['pickup_datetime']=pd.to_numeric(file['pickup_datetime'])
```

```python
file.info()
```

```python
file['key'] = file['key'].str.slice(0, 26)
file['key'] = pd.to_datetime(file['key'])
```

```python
file['key']=pd.to_numeric(file['key'])
```

```python
file
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
file[['key','pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']]=scaler.fit_transform(file[['key','pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']])
```

```python
file.describe()
```

```python
file[['fare_amount','key','pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']]
plt.figure(figsize=(10, 8))
sns.heatmap(file[['fare_amount','key','pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']]
.corr(), annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
labels=file['fare_amount']
features=file.drop(labels='fare_amount',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,labels,random_state=42,test_size=.3)
```

```python
from sklearn.linear_model import LinearRegression
LR_1=LinearRegression()
LR_1.fit(x_train,y_train)

```

```python
LR_1_pre=LR_1.predict(x_test)
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score
rmse=root_mean_squared_error(y_test,LR_1_pre)
LR_r2=r2_score(y_test,LR_1_pre)
print("rmse= ",rmse)
print("R2= ",LR_r2)
```

```python
from sklearn.svm import LinearSVR
model = LinearSVR(max_iter=5000)
model.fit(x_train, y_train)
```

```python
model_pre=model.predict(x_test)
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score
r2=r2_score(y_test,model_pre)
```

```python
print(r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import root_mean_squared_error,r2_score

model_GB=GradientBoostingRegressor(n_estimators=5,learning_rate=0.01,max_depth=10)

model_GB.fit(x_train,y_train)

model_GB_pre=model_GB.predict(x_test)

rmse_GB=root_mean_squared_error(y_test,model_GB_pre)
r2_GB=r2_score(y_test,model_GB_pre)

print('rmse= ',rmse_GB)
print('r2= ',r2_GB)

```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import root_mean_squared_error,r2_score

model_GB_2=GradientBoostingRegressor(n_estimators=50,learning_rate=0.01,max_depth=10)

model_GB_2.fit(x_train,y_train)

model_GB_pre_2=model_GB_2.predict(x_test)

rmse_GB_2=root_mean_squared_error(y_test,model_GB_pre_2)
r2_GB_2=r2_score(y_test,model_GB_pre_2)

print('rmse= ',rmse_GB_2)
print('r2= ',r2_GB_2)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import root_mean_squared_error,r2_score


model_AD=AdaBoostRegressor(n_estimators=70,learning_rate=0.01,random_state=42)

model_AD.fit(x_train,y_train)

model_AD_pre=model_AD.predict(x_test)

rmse_AD=root_mean_squared_error(y_test,model_AD_pre)
r2_AD=r2_score(y_test,model_AD_pre)

print('rmse= ',rmse_AD)
print('r2= ',r2_AD)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import root_mean_squared_error,r2_score

model_DT=DecisionTreeRegressor(max_depth=15,random_state=100,min_samples_leaf=10)

model_DT.fit(x_train,y_train)

model_DT_pre=model_DT.predict(x_test)

rmse_DT=root_mean_squared_error(y_test,model_DT_pre)
r2_DT=r2_score(y_test,model_DT_pre)

print('rmse= ',rmse_DT)
print('r2= ',r2_DT)
```
