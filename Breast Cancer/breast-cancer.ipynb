```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
file=pd.read_csv(r"breast-cancer-wisconsin-data_data.csv")
```

```python
## encode label (diagnosis)
file['diagnosis']=file['diagnosis'].map({'M':1,'B':0})
```

```python
import seaborn as sns

# Pair plot for multiple column relationships
sns.pairplot(file[['diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean']])
plt.show()
```

```python


fig, axes = plt.subplots(1, 2, figsize=(20, 10))  
 
sns.heatmap(file[['diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','concavity_se','compactness_se','smoothness_se','area_se','perimeter_se','radius_se','concave points_se','symmetry_se']].corr(), 
            annot=True, cmap='coolwarm', linewidths=0.5, ax=axes[0])
axes[0].set_title("Correlation Heatmap - First Set")

sns.heatmap(file[['diagnosis','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst','fractal_dimension_mean','symmetry_mean','radius_worst','fractal_dimension_se']].corr(), 
            annot=True, cmap='coolwarm', linewidths=0.5, ax=axes[1])
axes[1].set_title("Correlation Heatmap - Second Set")

plt.tight_layout()
plt.show()

```

```python
file.drop(labels=['Unnamed: 32','id'],inplace=True,axis=1)
```

```python
file.describe()
```

Which scaling method should you use?
Min-Max Scaling → If features have different ranges (e.g., [1, 1000] vs. [0.1, 0.9]).
Standardization (Z-score) → If data follows a normal distribution.
Robust Scaling → If data contains outliers.
Max Abs Scaling → If data has both positive and negative values but needs to stay within a small range.

```python
## make scaling for [ radius_mean , texture_mean , perimeter_mean , area_mean , radius_worst , texture_worst ,
#  perimeter_worst , area_worst , radius_se , texture_se,perimeter_se, area_se]
 
```

```python
## Using Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
file[['radius_mean' , 'texture_mean' , 'perimeter_mean' , 'area_mean' , 'radius_worst' , 'texture_worst']]=Scaler.fit_transform(file[['radius_mean' , 'texture_mean' , 'perimeter_mean' , 'area_mean' , 'radius_worst' , 'texture_worst']])
file[['perimeter_worst' , 'area_worst' , 'radius_se' , 'texture_se','perimeter_se', 'area_se']]=Scaler.fit_transform(file[['perimeter_worst' , 'area_worst' , 'radius_se' , 'texture_se','perimeter_se', 'area_se']])
```

```python
file.describe()
```

```python
## from hatemap we will take the features wich its relashionship with (diagnosis) is >= 0.5 as a features for training
labels=file['diagnosis']
features=file.drop(columns=['diagnosis','fractal_dimension_se','fractal_dimension_mean','symmetry_se','smoothness_se','compactness_se','concavity_se','texture_worst','smoothness_worst','symmetry_worst','fractal_dimension_worst',
'symmetry_mean','concave points_se','smoothness_mean','texture_mean'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.8,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model=LogisticRegression(max_iter=1000)
```

```python
logistic_model.fit(x_train,y_train)
```

```python
logistic_model_predictions=logistic_model.predict(x_test)
```

```python
logistic_model_predictions
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print(classification_report(y_test,logistic_model_predictions))

accuracy=accuracy_score(y_test,logistic_model_predictions)
print("Accuracy = ",accuracy)

confusion_matrix=confusion_matrix(y_test,logistic_model_predictions)
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not injured', 'injured'], yticklabels=['Not injured', 'injured'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
```

```python
##### Find the Best C, Gamma, and Degree

param_grid = {
    'C': [0.1, 1, 10, 100],  
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1], 
    'degree': [2, 3, 4, 5], 
    'kernel': ['rbf', 'poly']  
}

svm = SVC()

grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)  

print("Best Parameters:", grid_search.best_params_)

best_svm = grid_search.best_estimator_
accuracy = best_svm.score(x_test, y_test)
print("Test Accuracy:", accuracy)
```

```python
## Apply SVM
SVC_model=SVC(C=1,kernel='rbf',degree=2,gamma=1,max_iter=1000)
```

```python
SVC_model.fit(x_train,y_train)
```

```python
SVC_model_prediction=SVC_model.predict(x_test)
```

```python
SVC_model_prediction
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print(classification_report(y_test,SVC_model_prediction))

accuracy=accuracy_score(y_test,SVC_model_prediction)
print("Accuracy = ",accuracy)

confusion_matrix=confusion_matrix(y_test,SVC_model_prediction)
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not injured', 'injured'], yticklabels=['Not injured', 'injured'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
