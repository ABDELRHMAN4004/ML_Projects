# Make Encoding for All Object Data Type [ gender ,physical_Activity_Level,stress_Level,Chist_Pain_Type,Thalassemia,ECG_Results ]

```python
file['Gender']=file['Gender'].map({'Male':1,'Female':0})
```

```python
file['Physical_Activity_Level']=file['Physical_Activity_Level'].map(file['Physical_Activity_Level'].value_counts())
file['Stress_Level']=file['Stress_Level'].map(file['Stress_Level'].value_counts())
file['Chest_Pain_Type']=file['Chest_Pain_Type'].map(file['Chest_Pain_Type'].value_counts())
file['Thalassemia']=file['Thalassemia'].map(file['Thalassemia'].value_counts())
file['ECG_Results']=file['ECG_Results'].map(file['ECG_Results'].value_counts())
file['Heart_Attack_Risk']=file['Heart_Attack_Risk'].map({'Low':0,'Moderate':1,'High':2})
```

```python
file
```

# Make Scaling for columns

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

columns_scale=['Age','Physical_Activity_Level','BMI','Cholesterol_Level',
               'Resting_BP','Heart_Rate','Stress_Level','Chest_Pain_Type',
               'Thalassemia','ECG_Results','Max_Heart_Rate_Achieved',]
file[columns_scale]=scaler.fit_transform(file[columns_scale])
```

```python
labels=file['Heart_Attack_Risk']
features=file.drop(labels=['Fasting_Blood_Sugar','Age','BMI','Heart_Attack_Risk'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=.3,random_state=42)
```

# Apply Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model_1_LR = LogisticRegression(multi_class='ovr', solver='saga', max_iter=40000, class_weight='balanced')
model_1_LR.fit(x_train, y_train)
model_1_LR_prediction=model_1_LR.predict(x_test)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(classification_report(y_test,model_1_LR_prediction))
accuracy=accuracy_score(y_test,model_1_LR_prediction)
print("Accuracy = ",accuracy)

cm = confusion_matrix(y_test, model_1_LR_prediction)

labels = ['Low', 'Moderate', 'High']

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

# Apply Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model_1=DecisionTreeClassifier(criterion='log_loss',max_depth=5,min_samples_split=3,splitter='best',)

model_1.fit(x_train,y_train)

prediction=model_1.predict(x_test)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(classification_report(y_test,prediction))
accuracy=accuracy_score(y_test,prediction)
print("Accuracy = ",accuracy)

cm = confusion_matrix(y_test, prediction)

labels = ['Low', 'Moderate', 'High']

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn import tree

tree.plot_tree(model_1,label='root')
```

# Let Model Choose The Parameters

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

param_grid = {
    'criterion': ['gini', 'entropy'],  
    'max_depth': [3, 5, 10, None],  
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 5, 10],  
}

dt = DecisionTreeClassifier(random_state=42)

grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

print(f"Best Parameters : {grid_search.best_params_}")

best_tree = grid_search.best_estimator_
y_pred = best_tree.predict(x_test)

print(classification_report(y_test, y_pred))
```
