```python
import pandas as pd
import matplotlib.pyplot  as plt
import numpy as np
import seaborn as sns
```

```python
file=pd.read_csv("Airline Passenger Satisfaction.csv")
```

```python
file
```

```python
file.info()
```

```python
file.isnull().sum()
```

```python
file.dropna(axis=0,inplace=True)
```

```python
file.isnull().sum()
```

```python
file['Gender']=file['Gender'].map({'Male':1,'Female':0})
```

```python
file['Class'].value_counts()
```

```python
file['Class']=file['Class'].map({'Business':0,'Eco':1,'Eco Plus':-1})
```

```python
file['Customer Type'].value_counts()
```

```python
file['Customer Type']=file['Customer Type'].map({'Loyal Customer':0,'disloyal Customer':1})
```

```python
file['Type of Travel'].value_counts()
```

```python
file['Type of Travel']=file['Type of Travel'].map({'Business travel':0,'Personal Travel':1})
```

```python
file['satisfaction'].value_counts()
```

```python
file['satisfaction']=file['satisfaction'].map({'neutral or dissatisfied':0,'satisfied':1})
```

```python
file
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_cols = ['satisfaction', 'id', 'Unnamed: 0', 'Age']
df_selected = file[selected_cols]  

correlation_matrix = df_selected.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
file.drop(labels=['Unnamed: 0','id'],inplace=True,axis=1)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(file.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
file.drop(labels=['Gender','Departure/Arrival time convenient','Gate location','Departure Delay in Minutes','Arrival Delay in Minutes'],inplace=True,axis=1)
```

```python
file
```

```python
labels=file['satisfaction']
features=file.drop(labels='satisfaction',axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
columns_in_scale=['Age','Flight Distance','Inflight wifi service','Ease of Online booking','Food and drink',
                  'Online boarding','Seat comfort','Inflight entertainment','On-board service','Leg room service',
                  'Baggage handling','Checkin service','Inflight service','Cleanliness']
file[columns_in_scale]=scaler.fit_transform(file[columns_in_scale])
```

```python
file
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=.3,random_state=60)
```

```python
from sklearn.linear_model import LogisticRegression

model_1_logistic=LogisticRegression(C=.01,max_iter=2000)
model_1_logistic.fit(x_train,y_train)
```

```python
prediction_1=model_1_logistic.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Print Classification Report
print(classification_report(y_test, prediction_1))

# Calculate Accuracy
accuracy = accuracy_score(y_test, prediction_1)
print("Accuracy =", accuracy)

# Compute Confusion Matrix (Avoid overwriting function name)
conf_matrix = confusion_matrix(y_test, prediction_1)

# Plot Confusion Matrix Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['neutral or dissatisfied', 'satisfied'], 
            yticklabels=['neutral or dissatisfied', 'satisfied'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn import tree
model_t1=tree.DecisionTreeClassifier()
model_t1.fit(x_train,y_train)
prediction_t1=model_t1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Print Classification Report
print(classification_report(y_test, prediction_t1))

# Calculate Accuracy
accuracy = accuracy_score(y_test, prediction_t1)
print("Accuracy =", accuracy)

# Compute Confusion Matrix (Avoid overwriting function name)
conf_matrix = confusion_matrix(y_test, prediction_t1)

# Plot Confusion Matrix Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['neutral or dissatisfied', 'satisfied'], 
            yticklabels=['neutral or dissatisfied', 'satisfied'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

param_grid = {
    'criterion': ['gini', 'entropy'],  
    'max_depth': [3, 5, 10, None],  
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 5, 10],  
}

dt = DecisionTreeClassifier(random_state=42)

grid_search = GridSearchCV(dt, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

print(f"Best Parameters : {grid_search.best_params_}")

best_tree = grid_search.best_estimator_
y_pred = best_tree.predict(x_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
model_t2=tree.DecisionTreeClassifier(criterion='entropy',max_depth=None,min_samples_leaf=10,min_samples_split=2)
model_t2.fit(x_train,y_train)
prediction_t2=model_t2.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Print Classification Report
print(classification_report(y_test, prediction_t2))

# Calculate Accuracy
accuracy = accuracy_score(y_test, prediction_t2)
print("Accuracy =", accuracy)

# Compute Confusion Matrix (Avoid overwriting function name)
conf_matrix = confusion_matrix(y_test, prediction_t2)

# Plot Confusion Matrix Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['neutral or dissatisfied', 'satisfied'], 
            yticklabels=['neutral or dissatisfied', 'satisfied'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
