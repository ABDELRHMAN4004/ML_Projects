```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv("2017_Yellow_Taxi_Trip_Data.csv")
```

```python
file
```

```python
file.info()
```

```python
file.describe()
```

```python
file['tpep_dropoff_datetime']=pd.to_datetime(file['tpep_dropoff_datetime'])
file['tpep_pickup_datetime']=pd.to_datetime(file['tpep_pickup_datetime'])

file['duration']=file['tpep_pickup_datetime']-file['tpep_dropoff_datetime']

file.insert(1, column='duration', value=file.pop('duration'))

file.drop(labels=['tpep_pickup_datetime','tpep_dropoff_datetime'],inplace=True,axis=1)

file['duration']=pd.to_numeric(file['duration'])
```

```python
file['store_and_fwd_flag'].value_counts()
```

```python
file['store_and_fwd_flag']=file['store_and_fwd_flag'].map({'N':1,'Y':0})
```

```python
file.info()
```

```python
file
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
```

```python
file[['Unnamed: 0','duration','VendorID','passenger_count','trip_distance','RatecodeID','payment_type','extra','mta_tax','tolls_amount','PULocationID','DOLocationID','tip_amount','total_amount']]=Scaler.fit_transform(file[['Unnamed: 0','duration','VendorID','passenger_count','trip_distance','RatecodeID','payment_type','extra','mta_tax','tolls_amount','PULocationID','DOLocationID','tip_amount','total_amount']])
```

```python
file
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(file.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python

labels=file['fare_amount']
features=file.drop(labels=['fare_amount'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=.4,random_state=100) 
```

```python
from sklearn.linear_model import LinearRegression

model_LR_1=LinearRegression()
model_LR_1.fit(x_train,y_train)
prediction_1=model_LR_1.predict(x_test)
model_LR_1.intercept_
model_LR_1.coef_
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,prediction_1)
print(mse)
```

```python
plt.scatter(y_test,prediction_1)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual Vs Predicted")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,prediction_1)
print("R2 Score =",r2)
```

```python

errors = y_test -prediction_1
sns.histplot(errors, bins=50, kde=True)
plt.xlabel("Prediction Error")
plt.ylabel("Frequency")
plt.title("Distribution of Prediction Errors")
plt.show()
```

# Let's Apply Plonomial Regrassion

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

degree = 2
model_poly = make_pipeline(PolynomialFeatures(degree), LinearRegression())

model_poly.fit(x_train, y_train)
model_poly_predictions=model_poly.predict(x_test)
r2_poly = model_poly.score(x_test, y_test)
print("R² Score (Polynomial Regression):", r2_poly)
```

```python
plt.scatter(y_test,model_poly_predictions)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual Vs Predicted")
plt.show()
```

```python
from sklearn import svm
model_SVR_1=svm.SVR()
model_SVR_1.fit(x_train,y_train)
pre=model_SVR_1.predict(x_test)
r_2=model_SVR_1.score(x_test,y_test)
print("R2 score",r_2)
```
